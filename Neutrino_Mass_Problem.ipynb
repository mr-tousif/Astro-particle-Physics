Scale of Neutrino Masses:

For the neutrino masses, although we do not know the absolute mass scale, we
have the information on the squared-mass differences. Thus, it is possible to express the neutrino masses as functions of only one unknown parameter representing the absolute mass scale. It is convenient to take
this parameter as the value of the lightest mass, which is $m_1$ in the normal scheme
and $m_3$ in the inverted scheme.


Solar mass splitings: $$\Delta m_{21}^2=m_2^2-m_1^2$$
$$\Delta m_{21}=m_2-m_1$$

Atmosspheric mass splitings;$$ \Delta m_{31}^2=m_3^2-m_1^2$$
$$\Delta m_{31}=m_3-m_1$$

Mass Hierarchy:

NO: $m_1$ is the lightest  $\space$  $m_1<<m_2<<m_3$,  $\space$ $m_2^2=m_1^2+\Delta m_{21}^2$, $\space$   $m_3^2=m_1^2+\Delta m_{31}^2$

IO:  $m_3$ is the lightest  $\space$  $m_3<<m_1<<m_2$ and  $\space$ $m_1^2=m_3^2+\Delta m_{31}^2$, $\space$   $m_2^2=m_1^2+\Delta m_{21}^2=m_3^2+\Delta m_{31}^2+\Delta m_{21}^2$

```python
#Values of mass splitings from Super-K expt.
    
# Solar mass splitings in eV^2:

Delta_m_21_sqr=8.1*1e-5       # best fit
Delta_m_21_sqr_low=7.5*1e-5   #Lower bound 3sigma
Delta_m_21_sqr_high=8.7*1e-5  #Upper bound 3sigma


#Atmospheric Mass splitings in eV^2: 


Delta_m_31_sqr=2.2*1e-3       # best fit
Delta_m_31_sqr_low=1.4*1e-3    #Lower bound 3sigma
Delta_m_31_sqr_high=3.3*1e-3   #Upper bound 3sigma
```

```python
#Normal Hierachy: m_1<<m_2<<m_3

m_1=np.logspace(-4,0,5000)  #Lightest neutrino mass in eV

def m_2_NO(m_1, Delta_m_21_sqr):
    return np.sqrt(m_1**2+Delta_m_21_sqr)

def m_3_NO(m_1,Delta_m_31_sqr_NO):
    return np.sqrt(m_1**2+Delta_m_31_sqr)
```

```python
m_2_NO_bestfit=m_2_NO(m_1, Delta_m_21_sqr)
m_3_NO_bestfit=m_3_NO(m_1,Delta_m_31_sqr)

#3sigma fit

m_2_NO_low=m_2_NO(m_1, Delta_m_21_sqr_low)
m_2_NO_high=m_2_NO(m_1, Delta_m_21_sqr_high)

m_3_NO_low=m_3_NO(m_1,Delta_m_31_sqr_low)
m_3_NO_high=m_3_NO(m_1,Delta_m_31_sqr_high)
```

```python
#Inverted Hieracgy: m_3<<m_1=m_2

m_3=np.logspace(-4,0,5000)  #Lightest neutrino mass in eV

def m_1_IO(m_3,Delta_m_31_sqr):
    return np.sqrt(m_3**2+Delta_m_31_sqr)

def m_2_IO(m_3,Delta_m_31_sqr, Delta_m_21_sqr):
    return np.sqrt(m_3**2+Delta_m_31_sqr+Delta_m_21_sqr)
```

```python
m_1_IO_bestfit=m_1_IO(m_3,Delta_m_31_sqr)
m_2_IO_bestfit=m_2_IO(m_3,Delta_m_31_sqr,Delta_m_21_sqr)

#3sigma fit
m_1_IO_low=m_1_IO(m_3,Delta_m_31_sqr_low)
m_1_IO_high=m_1_IO(m_3,Delta_m_31_sqr_high)

m_2_IO_low=m_2_IO(m_3,Delta_m_31_sqr_low, Delta_m_21_sqr_low)
m_2_IO_high=m_2_IO(m_3,Delta_m_31_sqr_high, Delta_m_21_sqr_high)
```

Based on Oscillation Experiments: We see that in both schemes there is quasidegeneracy of the three masses when $m_1\approx m_2\approx m_3=m_\nu$ with, 

$$m_\nu\gg\sqrt{\Delta m_{Atm}^2}\approx5\times 10^{-2} eV$$ In this case, it is very difficult to distinguish the two schemes. On the other hand, the two schemes have very different features if the lightest mass is much smaller
than $\sqrt{\Delta m_{Atm}^2}$. In this case, in the normal scheme there is a hierarchy of masses: $m_1\ll m_2\ll m_3$. 

In the inverted scheme there is: $$m_3\ll m_1\approx m_2$$ in which $m_1$ and $m_2$ are quasidegenerate because their separation is due to the small solar squared-mass difference $\Delta m_{Sol}^2$. Finally we emphasize that, independently of the mass scheme, at least two neutrinos are massive, with masses larger than about $8\times 10^{-3}$ eV.
 

Assuming three Majorana neutrinos, $m_{ee}$ can be written in terms of the neutrino masses $m_i$, mixing angles  $\theta_{ij}$ and Majorana phases $\alpha_2$ and $\alpha_3$ as,


$$m_{ee}=\sum_{i=1}^{3} U_{ei}^{2} m_i=\cos^2\theta_{13}\cos^2\theta_{12} m_1+e^{i\alpha_2}\cos^2\theta_{13}\sin^2\theta_{12}  m_2 +e^{i\alpha_3}\sin^2\theta_{13}m_3$$


$$m_{ee}=\sum_{i=1}^{3} U_{ei}^{2} m_i=\left| U_{e1}\right|^2 m_1+e^{i\alpha_2}|U_{e2}|^2m_2+e^{i\alpha_3}|U_{e3}|^2m_3$$

We have $0\leq\alpha_k\leq 2\pi$, $k=2,3$. The two phases $α_2$
and $α_3$ we have the physically relevant cases: $\alpha_k=0,\pi$ for which CP is conserved: $e^{i\alpha_k}=\pm 1$ 

There are four possibilities:


(++) $\space$ $α_2 = α_3 = 0$ $\space$  ---> $\space$  $m_{ee}=\left| U_{e1}\right|^2 m_1+|U_{e2}|^2m_2+|U_{e3}|^2m_3$

$m_{ee}: $Maximum(++)

(--) $\space$ $α_2 = α_3 = \pi$ $\space$ ---> $\space$  $m_{ee}=\left| U_{e1}\right|^2 m_1-|U_{e2}|^2m_2-|U_{e3}|^2m_3$


(+-) $\space$ $α_2 =0, α_3 = \pi$ $\space$ ---> $\space$  $m_{ee}=\left| U_{e1}\right|^2 m_1+|U_{e2}|^2m_2-|U_{e3}|^2m_3$


(-+) $\space$ $α_2 =\pi, α_3 = 0$ $\space$ ---> $\space$  $m_{ee}=\left| U_{e1}\right|^2 m_1-|U_{e2}|^2m_2+|U_{e3}|^2m_3$


$m_{ee}$: Minimum(--,-+,+-)

Normal Order:

$$\boxed{m_{ee}=\left| U_{e1}\right|^2 m_1+e^{i\alpha_2}|U_{e2}|^2\sqrt{m_1^2+\Delta m_{21}^2}+e^{i\alpha_3}|U_{e3}|^2\sqrt{m_1^2+\Delta m_{31}^2}}$$


Inverted Order:

$$\boxed{m_{ee}=\left| U_{e1}\right|^2 \sqrt{m_3^2+\Delta m_{31}^2}+e^{i\alpha_2}|U_{e2}|^2\sqrt{m_1^2+\Delta m_{21}^2}+e^{i\alpha_3}|U_{e3}|^2m_3=\left| U_{e1}\right|^2 \sqrt{m_3^2+\Delta m_{31}^2}+e^{i\alpha_2}|U_{e2}|^2\sqrt{m_3^2+\Delta m_{31}^2+\Delta m_{21}^2}+e^{i\alpha_3}|U_{e3}|^2m_3}$$




It is straight forward to see in NO, we have $m_1\leq m_2\leq m_3$, and $m_2\approx \sqrt{\Delta m_{21}^2}=\sqrt{\Delta m_{Sol}^2}$ and $m_3\approx \sqrt{\Delta m_{31}^2}=\sqrt{\Delta m_{Atm}^2}$

In that scenario $m_{ee}$ can be written as, $\space$ ${m_{ee}}=\left| U_{e1}\right|^2 m_1+e^{i\alpha_2}|U_{e2}|^2\sqrt{\Delta m_{21}^2}+e^{i\alpha_3}|U_{e3}|^2\sqrt{\Delta m_{31}^2}$. 


In order to understand the behavior of $|m_{ee}|$ as a function of $m_1$, let us consider the best-fit values of the oscillation parameters. Since the best-fit value of $|U_{e3}|$ is zero, (i.e close to zero we can neglect it). Hence we are left with,

${m_{ee}}=\left| U_{e1}\right|^2 m_1+e^{i\alpha_2}|U_{e2}|^2\sqrt{\Delta m_{21}^2}$

Special case:

(--) $\space$ $α_2 = α_3 = \pi$ $\space$---> ${m_{ee}}=\left| U_{e1}\right|^2 m_1-|U_{e2}|^2\sqrt{\Delta m_{21}^2}$
    
(-+) $\space$ $α_2 = \pi,α_3 = 0$ $\space$ ---> ${m_{ee}}=\left| U_{e1}\right|^2 m_1-|U_{e2}|^2\sqrt{\Delta m_{21}^2}$

It is possible to have a vanishing $|m_{ee}|$ in the CP-conserving cases (--) and (-+) if,

$0=\left| U_{e1}\right|^2 m_1-|U_{e2}|^2\sqrt{\Delta m_{21}^2}$ $\space$ ---> $m_1=\frac{|U_{e2}|^2\sqrt{\Delta m_{21}^2}}{\left| U_{e1}\right|^2}\approx 4\times 10^{-3}$ eV.


For smaller values of $m_1$, we can approximate $m_{ee}$

$|m_{ee}|\approx||U_{e2}|^2\sqrt{\Delta m_{21}^2} + e^{i(\alpha_3−\alpha_2)} |U_{e3}|^2\sqrt{\Delta m_{31}^2}|$

Since the best fit values of $U_{e3}$ is close to zero, we have $|m_{ee}|\approx||U_{e2}|^2\sqrt{\Delta m_{21}^2}|$

    


```python
#Normal order:NO


def abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr,Delta_m_31_sqr, U_e1,U_e2,U_e3):  #effective neutrino mass
    factor_1=np.sqrt(m_1**2+Delta_m_21_sqr)
    factor_2=np.sqrt(m_1**2+Delta_m_31_sqr)
    m_ee=U_e1**2*m_1+np.exp(1j*alpha_2)*U_e2**2*factor_1+np.exp(1j*alpha_3)*U_e3**2*factor_2
    #m_ee_conj=np.conj(m_ee)
    return np.abs(m_ee)  #absolute values


# Special case:

def neutrino_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr,Delta_m_31_sqr, U_e1,U_e2,U_e3):  #effective neutrino mass
    factor_1=np.sqrt(Delta_m_21_sqr)
    factor_2=np.sqrt(Delta_m_31_sqr)
    m_ee=U_e1**2*m_1+np.exp(1j*alpha_2)*U_e2**2*factor_1+np.exp(1j*alpha_3)*U_e3**2*factor_2
    m_ee_conj=np.conj(m_ee)
    return np.abs(m_ee)  #absolute values
```

```python
# (++) 

alpha_2=0
alpha_3=0   #l.b=lower bound, u.b=upper bound

m_ee_NO_1=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr,Delta_m_31_sqr,U_e1,U_e2,U_e3) #best fit
m_ee_NO_1_l=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr_low,Delta_m_31_sqr_low,U_e1_low,U_e2_low,U_e3_low) #3sigma l.b
m_ee_NO_1_h=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr_high,Delta_m_31_sqr_high,U_e1_high,U_e2_high,U_e3_high) #3sigma u.b




#*******************************************************************************************************************

# (--) 

alpha_2=np.pi
alpha_3=np.pi

m_ee_NO_2=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr,Delta_m_31_sqr,U_e1,U_e2,U_e3) #best fit
m_ee_NO_2_l=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr_low,Delta_m_31_sqr_low,U_e1_low,U_e2_low,U_e3_low)#3sigma l.b
m_ee_NO_2_h=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr_high,Delta_m_31_sqr_high,U_e1_high,U_e2_high,U_e3_high)#3sigma u.b


#Special case:


mass_2=neutrino_NO(m_1,np.pi,np.pi,Delta_m_21_sqr,Delta_m_31_sqr, U_e1,U_e2,U_e3)
mass_2_l=neutrino_NO(m_1,np.pi,np.pi,Delta_m_21_sqr_low,Delta_m_31_sqr_low, U_e1_low,U_e2_low,U_e3_low)
mass_2_h=neutrino_NO(m_1,np.pi,np.pi,Delta_m_21_sqr_high,Delta_m_31_sqr_high, U_e1_low,U_e2_high,U_e3_high)

#*******************************************************************************************************************

# (+-) 
alpha_2=0
alpha_3=np.pi


m_ee_NO_3=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr,Delta_m_31_sqr,U_e1,U_e2,U_e3) #best fit
m_ee_NO_3_l=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr_low,Delta_m_31_sqr_low,U_e1_low,U_e2_low,U_e3_low) #3sigma l.b  
m_ee_NO_3_h=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr_high,Delta_m_31_sqr_high,U_e1_high,U_e2_high,U_e3_high)#3sigma u.b

#Special case:

mass_3=neutrino_NO(m_1,0,np.pi,Delta_m_21_sqr,Delta_m_31_sqr, U_e1,U_e2,U_e3)
mass_3_l=neutrino_NO(m_1,0,np.pi,Delta_m_21_sqr_low,Delta_m_31_sqr_low, U_e1_low,U_e2_low,U_e3_low)
mass_3_h=neutrino_NO(m_1,0,np.pi,Delta_m_21_sqr_high,Delta_m_31_sqr_high, U_e1_low,U_e2_high,U_e3_high)


#*******************************************************************************************************************

# (-+) 
alpha_2=np.pi
alpha_3=0


m_ee_NO_4=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr,Delta_m_31_sqr,U_e1,U_e2,U_e3) #best fit
m_ee_NO_4_l=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr_low,Delta_m_31_sqr_low,U_e1_low,U_e2_low,U_e3_low)#3sigma l.b
m_ee_NO_4_h=abs_m_ee_NO(m_1,alpha_2,alpha_3,Delta_m_21_sqr_high,Delta_m_31_sqr_high,U_e1_high,U_e2_high,U_e3_high)#3sigma u.b


#Special case:

mass_4=neutrino_NO(m_1,np.pi,0,Delta_m_21_sqr,Delta_m_31_sqr, U_e1,U_e2,U_e3)
mass_4_l=neutrino_NO(m_1,np.pi,0,Delta_m_21_sqr_low,Delta_m_31_sqr_low, U_e1_low,U_e2_low,U_e3_low)
mass_4_h=neutrino_NO(m_1,np.pi,0,Delta_m_21_sqr_high,Delta_m_31_sqr_high, U_e1_low,U_e2_high,U_e3_high)


```

```python

#Inverted order:IO


def abs_m_ee_IO(m_3,alpha_2,alpha_3,Delta_m_21_sqr,Delta_m_31_sqr, U_e1,U_e2,U_e3):  #effective neutrino mass
    factor_1=np.sqrt(m_3**2+Delta_m_31_sqr)
    factor_2=np.sqrt(m_3**2+Delta_m_31_sqr+Delta_m_21_sqr)
    m_ee=U_e1**2*factor_1+np.exp(1j*alpha_2)*U_e2**2*factor_2+np.exp(1j*alpha_3)*U_e3**2*m_3
    return np.abs(m_ee)  #absolute values


```

```python
#************************************************************************************************************

#Effective Neutrino Mass from Titrium beta decay as following:
```


NO: $m_1$ is the lightest  $\space$  $m_1<<m_2<<m_3$,  $\space$ $m_2^2=m_1^2+\Delta m_{21}^2$, $\space$   $m_3^2=m_1^2+\Delta m_{31}^2$


$$\boxed{m_\beta^2=c_{12}^2 c_{13}^2 m_1^2+s_{12}^2 c_{13}^2 m_{2}^2+s_{13}^2 m_{3}^2=c_{12}^2 c_{13}^2 m_1^2+s_{12}^2 c_{13}^2(m_1^2+\Delta m_{21}^2)+s_{13}^2(m_1^2+\Delta m_{31}^2)}$$


IO:  $m_3$ is the lightest  $\space$  $m_3<<m_1\approx m_2$ and  $\space$ $m_1^2=m_3^2+\Delta m_{31}^2$, $\space$ $m_2^2=m_1^2+\Delta m_{21}^2=m_3^2+\Delta m_{31}^2+\Delta m_{21}^2$


$$\boxed{m_\beta^2=c_{12}^2 c_{13}^2 m_1^2+s_{12}^2 c_{13}^2 m_{2}^2+s_{13}^2 m_{3}^2=c_{12}^2 c_{13}^2 (m_3^2+\Delta m_{31}^2)+s_{12}^2 c_{13}^2(m_3^2+\Delta m_{31}^2+\Delta m_{21}^2)+s_{13}^2 m_{3}^2}$$



```python
#best fit
s_12_sqr=0.30
s_13_sqr=0.0

#3sigma

s_12_sqr_low=0.23
s_13_sqr_low=0.0227

s_12_sqr_high=0.38
s_13_sqr_high=0.0247

#c_12=np.cos(theta_12)

c_12_sqr=1-s_12_sqr
#c_13=np.cos(theta_13)
c_13_sqr=1-s_13_sqr

#s_12=np.sin(theta_12)
#s_13=np.sin(theta_13)

def mass_NO(s_12_sqr,s_13_sqr,m_1,Delta_m_21_sqr,Delta_m_31_sqr): #Normal order
    m_beta_sqr=c_12_sqr*c_13_sqr*m_1**2+s_12_sqr*c_13_sqr*(m_1**2+Delta_m_21_sqr)+s_13_sqr*(m_1**2+Delta_m_31_sqr)
    return np.sqrt(m_beta_sqr)

def mass_IO(s_12_sqr,s_13_sqr,m_3, Delta_m_21_sqr,Delta_m_31_sqr):  #inverted order
    m_beta_sqr=c_12_sqr*c_13_sqr*(m_3**2+Delta_m_31_sqr)+s_12_sqr*c_13_sqr*(m_3**2+Delta_m_31_sqr+Delta_m_21_sqr)+s_13_sqr*m_3**2
    return np.sqrt(m_beta_sqr)


    
```

```python
y_1=mass_NO(s_12_sqr,s_13_sqr,m_1,Delta_m_21_sqr,Delta_m_31_sqr)
y_1l=mass_NO(s_12_sqr_low,s_13_sqr_low,m_1,Delta_m_21_sqr_low,Delta_m_31_sqr_low)
y_1h=mass_NO(s_12_sqr_high,s_13_sqr_high,m_1,Delta_m_21_sqr_high,Delta_m_31_sqr_high)


y_2=mass_IO(s_12_sqr,s_13_sqr,m_3, Delta_m_21_sqr,Delta_m_31_sqr)
y_2l=mass_IO(s_12_sqr_low,s_13_sqr_low,m_3, Delta_m_21_sqr_low,Delta_m_31_sqr_low)
y_2h=mass_IO(s_12_sqr_high,s_13_sqr_high,m_3, Delta_m_21_sqr_high,Delta_m_31_sqr_high)
```

```python
#**************************************************

#Neutrino mass Cosmology as following:
```

$$m_{cosmo}=m_1+m_2+m_3$$

NO: $m_1$ is the lightest  $\space$  $m_1<<m_2<<m_3$,  $\space$ $m_2^2=m_1^2+\Delta m_{21}^2$, $\space$   $m_3^2=m_1^2+\Delta m_{31}^2$ then $m_{cosmo}$ becomes,

$$m_{cosmo}=m_1+\sqrt{m_1^2+\Delta m_{21}^2}+\sqrt{m_1^2+\Delta m_{31}^2}$$


IO:  $m_3$ is the lightest  $\space$  $m_3<<m_1\approx m_2$ and  $\space$ $m_1^2=m_3^2+\Delta m_{31}^2$, $\space$ $m_2^2=m_1^2+\Delta m_{21}^2=m_3^2+\Delta m_{31}^2+\Delta m_{21}^2$ then $m_{cosmo}$ becomes,

$$m_{cosmo}=\sqrt{m_3^2+\Delta m_{31}^2}+\sqrt{m_3^2+\Delta m_{31}^2+\Delta m_{21}^2}+m_3$$

```python
def mass_cosmo_NO(m_1,Delta_m_21_sqr,Delta_m_31_sqr):  #Normal Order
    return m_1+np.sqrt(m_1**2+Delta_m_21_sqr)+np.sqrt(m_1**2+Delta_m_31_sqr)


def mass_cosmo_IO(m_3,Delta_m_21_sqr,Delta_m_31_sqr):  #Inverted Order
    return np.sqrt(m_3**2+Delta_m_31_sqr)+np.sqrt(m_3**2+Delta_m_31_sqr+Delta_m_21_sqr)+m_3
```

```python
#inverted order

cosmo_IO_bestfit=mass_cosmo_IO(m_3,Delta_m_21_sqr,Delta_m_31_sqr) #best fit
cosmo_IO_low=mass_cosmo_IO(m_3,Delta_m_21_sqr_low,Delta_m_31_sqr_low) #3sigma
cosmo_IO_high=mass_cosmo_IO(m_3,Delta_m_21_sqr_high,Delta_m_31_sqr_high) #3sigma
```
